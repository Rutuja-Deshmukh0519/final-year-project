```python
# Import NMF
from sklearn.decomposition import NMF
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import os 
import glob
import xlsxwriter
import math  
import time
%matplotlib inline 
```

```python
total_time = 0.0
sum_fit = 0.0
sum_transform = 0.0
start_total = time.process_time()
i = 0

os.mkdir('nmf-output-60-comps')
path="resized-output/*.jpeg*"

for image in glob.glob(path):
    img=io.imread(image,as_gray=True)

    start_fit = time.process_time()
    fit_nmf = NMF(n_components=60).fit(img)
    sum_fit += time.process_time() - start_fit
    
    start_transform = time.process_time()
    transformed = fit_nmf.transform(img)
    sum_transform += time.process_time() - start_transform
    
    projected = fit_nmf.inverse_transform(transformed)
    projected = projected.astype(np.uint8)
    # Print the NMF features
#     print(projected)

    #print(model.components_.shape)
#     plt.imshow(projected, cmap = 'gray')
                  
    io.imsave("nmf-output-60-comps/image%05i.jpeg" %i, projected)
    i += 1

    total_time += time.process_time() - start_total

avg_fit = sum_fit/1000;
avg_transform = sum_transform/1000

print("Total Fit Time: ", sum_fit)
print("Average Fit time: ", avg_fit)
print("Total Transform Time: ", sum_transform)
print("Average Transform time: ", avg_transform)
print("Total Time required for execution: ", total_time)
```

```python
def mse(imageA, imageB):
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err
```

```python
workbook = xlsxwriter.Workbook('NMF-Dimensionality-reduction-analysis.xlsx')
worksheet = workbook.add_worksheet()

worksheet.write('A1', 'log(MSE-NMF-40-components)')
worksheet.write('B1', 'log(MSE-NMF-45-components)')
worksheet.write('C1', 'log(MSE-NMF-50-components)')
worksheet.write('D1', 'log(MSE-NMF-55-components)')
worksheet.write('E1', 'log(MSE-NMF-60-components)')

path_resized = "D:/Documents/final-year-project/resized-output/*.jpeg*"
path_nmf_40_comps = "D:/Documents/final-year-project/nmf-output-40-comps/*.jpeg*"
path_nmf_45_comps = "D:/Documents/final-year-project/nmf-output-45-comps/*.jpeg*"
path_nmf_50_comps = "D:/Documents/final-year-project/nmf-output-50-comps/*.jpeg*"
path_nmf_55_comps = "D:/Documents/final-year-project/nmf-output-55-comps/*.jpeg*"
path_nmf_60_comps = "D:/Documents/final-year-project/nmf-output-60-comps/*.jpeg*"

rowIndexNMF = 2
for imgResized, imgNMF40Comps, imgNMF45Comps, imgNMF50Comps, imgNMF55Comps, imgNMF60Comps in zip(glob.glob(path_resized), glob.glob(path_nmf_40_comps), glob.glob(path_nmf_45_comps), glob.glob(path_nmf_50_comps), glob.glob(path_nmf_55_comps), glob.glob(path_nmf_60_comps)):
    img_resized = io.imread(imgResized)
    img_nmf_40_comps = io.imread(imgNMF40Comps)
    img_nmf_45_comps = io.imread(imgNMF45Comps)
    img_nmf_50_comps = io.imread(imgNMF50Comps)
    img_nmf_55_comps = io.imread(imgNMF55Comps)
    img_nmf_60_comps = io.imread(imgNMF60Comps)
    
    err_nmf_40_comps = mse(img_resized, img_nmf_40_comps)
    err_nmf_45_comps = mse(img_resized, img_nmf_45_comps)
    err_nmf_50_comps = mse(img_resized, img_nmf_50_comps)
    err_nmf_55_comps = mse(img_resized, img_nmf_55_comps)
    err_nmf_60_comps = mse(img_resized, img_nmf_60_comps)
    
    # Calculating log for better visualization
    log_mse_nmf_40_comps = math.log(err_nmf_40_comps)
    log_mse_nmf_45_comps = math.log(err_nmf_45_comps)
    log_mse_nmf_50_comps = math.log(err_nmf_50_comps)
    log_mse_nmf_55_comps = math.log(err_nmf_55_comps)
    log_mse_nmf_60_comps = math.log(err_nmf_60_comps)
    
    worksheet.write('A' + str(rowIndexNMF), log_mse_nmf_40_comps)
    worksheet.write('B' + str(rowIndexNMF), log_mse_nmf_45_comps)
    worksheet.write('C' + str(rowIndexNMF), log_mse_nmf_50_comps)
    worksheet.write('D' + str(rowIndexNMF), log_mse_nmf_55_comps)
    worksheet.write('E' + str(rowIndexNMF), log_mse_nmf_60_comps)
   
    rowIndexNMF += 1
    
workbook.close()

print("MSE Calculation Successful!")
```
